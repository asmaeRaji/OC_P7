# Modélisation 

Nous avons, comme vu avec nos graphs d'exploration, un grand problème de déséquilibre du jeu de données. Nous avons 92% de TARGET négative et donc 8% de TARGET positive. Cela pose problème à plusieurs échelles :

- Premièrement, lors de nos tests pour trouver les meilleurs hyperparamètres de chaque modèle, nous utiliserons une Cross Validation. Or, lorsqu'on utlisera des K-Fold, nous avons beaucoup de chances qu'un de nos fold ne contienne pas ou presque pas de TARGET positive, ce qui fera que l'entraînement sera très biaisé et prédira probablement trop de Négatif. Cela est plutôt grâve car on ne souhaite absolument pas avoir de Faux Négatifs (du point de vue de la banque). Cela voudrait dire accepter un dossier qu'on devrait refuser.

- Deuxièmement, au niveau des résultats, si l'on décide de choisir une metric cible pour les modèles du type "précision, accuracy", il suffirait à un modèle de prédire tout le temps une TARGET négative pour obtenir un bon score. On doit donc trouver une autre metric pour classer nos modèles. Une possibilité serait d'utiliser l'AUC avec, au lieu de prédire si un individu est positif ou négatif, une probabilité que l'individu soit positif.

- Troisièmement, au niveau du modèle, on souhaite que l'impact d'un faux négatif (prêt non remboursé) soit beaucoup plus grand qu'un faux positif (prêt non accordé qui aurait été remboursé) car cela coûte beaucoup plus à la banque. Combien de fois plus grand ? Pour l'instant on pourrait dire 10 fois pour vraiment faire un impact.

# Création des fonctions de test et de Cross Validation

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score
```

```python
def getCVParams(model_name):
    if(model_name == 'LogisticRegression'):
        #Il n'y a pas vraiment d'hyperparamètres à tune avec la logistic regression
        return {}
    if(model_name == 'DecisionTreeClassifier'):
        return {'max_depth': [5,10,50],
                'min_samples_split': [2,10],
                'min_samples_leaf':[1,10],
                'max_features':['auto', 'sqrt']}
    if(model_name == 'XGBClassifier'):
         return {'eta': [0.01, 0.2],
                'gamma': [1,10],
                'max_depth': [3,5,10]}
    if(model_name == 'LGBMClassifier'):
        return {
            'num_leaves': [25,100,150],
            'min_data_in_leaf' : [100,1000,5000],
            'max_depth': [5,10,15]
        }
```

```python
def getModel(model_name):
    if(model_name == 'LogisticRegression'):
        return LogisticRegression(class_weight = class_weight)
    if(model_name == 'DecisionTreeClassifier'):
        return DecisionTreeClassifier(class_weight=class_weight)
    if(model_name == 'XGBClassifier'):
         return XGBClassifier(scale_pos_weight=10)
    if(model_name == 'LGBMClassifier'):
        return LGBMClassifier(class_weight=class_weight)
```

```python
def getCrossValScores(model_name, X_train, y_train, X_test, y_test, table, nb_splits=10, nb_repeats=3):
    cvParams = getCVParams(model_name)
    model = getModel(model_name)
    cv = RepeatedStratifiedKFold(n_splits=nb_splits, n_repeats=nb_repeats, random_state=1)
    clf = GridSearchCV(estimator=model, param_grid=cvParams, cv=cv, scoring='roc_auc', verbose = 2)
    clf.fit(X_train, y_train)
    
    #Second part, to test with the test values
    best_params = clf.best_params_
    model.set_params(**best_params)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    test_score = roc_auc_score(y_test,y_pred)
    table.loc[len(table)] =[model_name, clf.best_score_, np.mean(clf.cv_results_['mean_fit_time']), np.mean(clf.cv_results_['mean_score_time']), 
                             test_score]
    print(table)
    return clf.cv_results_
```

## Clarifications sur notre but

On possède une classe y (TARGET) qui nous indique si un prêt que la banque à accordé a été remboursé ou non. Si cette colonne est égale à 1, alors le client à des problèmes de paiement (ne rembourse pas son prêt) comme l'indique la description:

```python
hc_columns_description.loc[1,'Description']
```

La banque souhaite donc minimiser le nombre de clients pour lequel elle aura prévu 0 (donc négatif, mais économiquement positif pour elle) et qu'il s'agira en fait d'un client qui donnera un 1 (donc positif).
Pour résumer, la banque souhaite donc fortement minimiser les faux négatifs.

Pour cela, le Recall (Vrais positifs/(Vrais Positifs + Faux Négatifs)) nous donne une première mesure qui serait peut-être plus à même d'insister sur ce point qu'un AUC ROC score.

On peut aussi tester, plutôt qu'un f1Score, un fBetaScore qui prendrait plus en compte le recall. Le f1 score est la moyenne harmonique de la précision et du recall, mais le fBetaScore a un paramètre ajustable qui nous laisse indiquer si l'un est plus important que l'autre. Ici, on peut aussi utiliser le chiffre 10 pour dire qu'on veut prendre en compte 10x plus le recall que la précision.

```python
yhat = final_model.predict_proba(X_test)
```

```python
probs = yhat[:, 1]
```

```python
# define thresholds
thresholds = arange(0, 1, 0.001)
```

```python
# apply threshold to positive probabilities to create labels
def to_labels(pos_probs, threshold):
    return (pos_probs >= threshold).astype('int')
```

```python
# evaluate each threshold
f10_scores = [fbeta_score(y_test, to_labels(probs, t), beta=2.5) for t in thresholds]
roc_auc_scores = [roc_auc_score(y_test, to_labels(probs,t)) for t in thresholds]
accuracy_scores = [accuracy_score(y_test, to_labels(probs,t)) for t in thresholds]
f_10_and_acc_scores = np.add(f10_scores,accuracy_scores)
```

```python
# get best threshold
f_10ix = argmax(f10_scores)
auc_ix = argmax(roc_auc_scores)
acc_ix = argmax(accuracy_scores)
mixed_ix = argmax(f_10_and_acc_scores)
print('Threshold=%.3f, F10-Score=%.5f, ROCAUC Score=%.5f, Accuracy=%.5f, Mixed=%5f' % (thresholds[f_10ix], f10_scores[f_10ix], roc_auc_scores[f_10ix], accuracy_scores[f_10ix], f_10_and_acc_scores[f_10ix]))
print('Threshold=%.3f, F10-Score=%.5f, ROCAUC Score=%.5f, Accuracy=%.5f, Mixed=%5f' % (thresholds[auc_ix], f10_scores[auc_ix], roc_auc_scores[auc_ix], accuracy_scores[auc_ix], f_10_and_acc_scores[auc_ix]))
print('Threshold=%.3f, F10-Score=%.5f, ROCAUC Score=%.5f, Accuracy=%.5f, Mixed=%5f' % (thresholds[acc_ix], f10_scores[acc_ix], roc_auc_scores[acc_ix], accuracy_scores[acc_ix], f_10_and_acc_scores[acc_ix]))
print('Threshold=%.3f, F10-Score=%.5f, ROCAUC Score=%.5f, Accuracy=%.5f, Mixed=%5f' % (thresholds[acc_ix], f10_scores[acc_ix], roc_auc_scores[acc_ix], accuracy_scores[acc_ix], f_10_and_acc_scores[acc_ix]))
print('Threshold=%.3f, F10-Score=%.5f, ROCAUC Score=%.5f, Accuracy=%.5f, Mixed=%5f' % (thresholds[mixed_ix], f10_scores[mixed_ix], roc_auc_scores[mixed_ix], accuracy_scores[mixed_ix], f_10_and_acc_scores[mixed_ix]))
```
